## Cargar y preparar los datos

```python
import pandas as pd
import matplotlib.pyplot as plt


# Cargar los datos
visits = pd.read_csv('/datasets/visits_log_us.csv', parse_dates=['Start Ts', 'End Ts'])
orders = pd.read_csv('/datasets/orders_log_us.csv', parse_dates=['Buy Ts'])
costs = pd.read_csv('/datasets/costs_us.csv', parse_dates=['dt'])

# Verificar las primeras filas de cada dataset
visits.head(), orders.head(), costs.head()
```

## Convertir tipos de Datos

```python
# Convertir columnas de fecha a tipo datetime
visits['Start Ts'] = pd.to_datetime(visits['Start Ts'])
visits['End Ts'] = pd.to_datetime(visits['End Ts'])
orders['Buy Ts'] = pd.to_datetime(orders['Buy Ts'])
costs['dt'] = pd.to_datetime(costs['dt'])

# Verificar tipos de datos
visits.dtypes, orders.dtypes, costs.dtypes
```

<div class="alert alert-block alert-success"> <b>Comentario del revisor</b> <a class="tocSkip"></a><br>

<b>Éxito</b> - ¡Buen trabajo asegurándote de que las columnas de fecha estén correctamente convertidas al tipo datetime! Esto es esencial para realizar análisis basados en tiempo, como agrupaciones o cálculos de intervalos. La validación adicional mediante dtypes es una práctica muy recomendable. ¡Sigue así! 😊
</div>


## Análisis de las Visitas

### ¿Cuántas personas usan el servicio cada día, semana y mes?

```python
# Visitas diarias
visitas_diarias = visits.groupby(visits['Start Ts'].dt.date)['Uid'].nunique()

# Visitas semanales
visitas_semanales = visits.groupby(visits['Start Ts'].dt.to_period('W'))['Uid'].nunique()

# Visitas mensuales
visitas_mensuales = visits.groupby(visits['Start Ts'].dt.to_period('M'))['Uid'].nunique()

# Mostrar resultados
visitas_diarias, visitas_semanales, visitas_mensuales
```

<div class="alert alert-block alert-success"> <b>Comentario del revisor</b> <a class="tocSkip"></a><br>

<b>Éxito</b> - ¡Excelente análisis! Agrupar los datos de visitas por día, semana y mes proporciona una perspectiva muy clara sobre el comportamiento de los usuarios en diferentes intervalos de tiempo. Además, el uso de nunique para identificar usuarios únicos es una elección adecuada para este tipo de análisis. ¡Buen trabajo! 🚀
</div>

#### Gráfico de visitas

```python
# Graficar visitas diarias
plt.figure(figsize=(10,6))
plt.plot(visitas_diarias.index, visitas_diarias.values)
plt.title('Visitas Diarias')
plt.xlabel('Fecha')
plt.ylabel('Número de visitas')
plt.xticks(rotation=45)
plt.show()
```

<div class="alert alert-block alert-success"> <b>Comentario del revisor</b> <a class="tocSkip"></a><br>

<b>Éxito</b> - ¡Gran visualización! El gráfico de visitas diarias es claro y fácil de interpretar. La elección de rotar las etiquetas del eje X mejora la legibilidad, especialmente para conjuntos de datos temporales. Esto ayudará a identificar patrones o tendencias en el comportamiento de los usuarios a lo largo del tiempo. 

### ¿Cuántas sesiones hay por día?

```python
# Número de sesiones por día
sesiones_diarias = visits.groupby(visits['Start Ts'].dt.date)['Uid'].count()
sesiones_diarias
```

<div class="alert alert-block alert-success"> <b>Comentario del revisor</b> <a class="tocSkip"></a><br>

<b>Éxito</b> - ¡Buen trabajo al calcular el número de sesiones diarias! Los resultados son claros y muestran la cantidad de sesiones agrupadas por fecha. Esta información será clave para identificar tendencias o patrones en la actividad del usuario. Muy bien estructurado, ¡continúa así! 🚀📊

### ¿Cuál es la duración de cada sesión?

```python
# Calcular duración de la sesión en minutos
visits['session_duration'] = (visits['End Ts'] - visits['Start Ts']).dt.total_seconds() / 60

# Duración promedio por sesión
duracion_promedio = visits.groupby(visits['Start Ts'].dt.date)['session_duration'].mean()
duracion_promedio 

# Graficar la duración promedio por sesión
plt.figure(figsize=(10, 6))
plt.plot(duracion_promedio.index, duracion_promedio.values, marker='o', color='b', linestyle='-', linewidth=2, markersize=5)
plt.title('Duración Promedio de la Sesión por Día', fontsize=14)
plt.xlabel('Fecha', fontsize=12)
plt.ylabel('Duración Promedio (minutos)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

Se observa que en promedio las visitas duran alrededor de 10 minutos

<div class="alert alert-block alert-success"> <b>Comentario del revisor</b> <a class="tocSkip"></a><br>

<b>Éxito</b> - Excelente análisis de la duración promedio de las sesiones. Has calculado de manera precisa la duración de cada sesión y la has visualizado claramente con un gráfico que facilita la interpretación de las tendencias diarias. La conclusión sobre la duración promedio es clara y bien fundamentada. ¡Sigue adelante, estás haciendo un gran trabajo! 🌟📈
</div>

## Análisis de las Ventas

### ¿Cuándo empieza la gente a comprar?

```python
# Unir la tabla de visitas con la de pedidos para obtener la primera compra de cada usuario
visitas_con_compras = visits.merge(orders[['Uid', 'Buy Ts']], on='Uid', how='left')
visitas_con_compras['time_to_first_purchase'] = (visitas_con_compras['Buy Ts'] - visitas_con_compras['Start Ts']).dt.days

# Mostrar la distribución de días hasta la primera compra
visitas_con_compras['time_to_first_purchase'].describe()
```

La mayoría (el 75% de los usuarios) realiza su primera compra luego de 51 días de su primera visita.

### ¿Cuánto dinero traen los usuarios? (LTV)

```python
# Calcular el LTV por usuario
ltv_por_usuario = orders.groupby('Uid')['Revenue'].sum()

# Promedio de LTV
ltv_promedio = ltv_por_usuario.mean()
ltv_promedio
```

## Análisis de los Costos de Marketing

### ¿Cuánto dinero se gastó? (Total, por fuente de adquisición, a lo largo del tiempo)

```python
# Gastos totales por fecha
gastos_totales = costs.groupby('dt')['costs'].sum()

# Gastos por fuente
gastos_por_fuente = costs.groupby('source_id')['costs'].sum()

# Gastos totales a lo largo del tiempo
gastos_totales

# Graficar los gastos totales a lo largo del tiempo
plt.figure(figsize=(10, 6))
plt.plot(gastos_totales.index, gastos_totales.values, marker='o', color='b', linestyle='-', linewidth=2, markersize=5)
plt.title('Gastos Totales por Fecha', fontsize=14)
plt.xlabel('Fecha', fontsize=12)
plt.ylabel('Gastos Totales', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# Graficar los gastos por fuente
plt.figure(figsize=(10, 6))
gastos_por_fuente.plot(kind='bar', color='orange')
plt.title('Gastos por Fuente', fontsize=14)
plt.xlabel('Fuente', fontsize=12)
plt.ylabel('Gastos Totales', fontsize=12)
plt.xticks(rotation=0)
plt.grid(True)
plt.tight_layout()
plt.show()
```

<div class="alert alert-block alert-success"> <b>Comentario del revisor</b> <a class="tocSkip"></a><br>

<b>Éxito</b> - ¡Buen trabajo al analizar los costos de marketing! Visualizar los gastos totales a lo largo del tiempo y por fuente de adquisición es un excelente enfoque para identificar patrones y evaluar la efectividad de las campañas. Las gráficas están claras y bien presentadas, lo que facilita la interpretación de los resultados. ¡Continúa con este nivel de detalle y claridad! 📊✨

### ¿Cuál fue el costo de adquisición de clientes de cada una de las fuentes?

```python
# Unir costos con pedidos por fuente
pedidos_por_fuente = orders.merge(visits[['Uid', 'Source Id']], on='Uid', how='left')

# Calcular CAC
clientes_por_fuente = pedidos_por_fuente.groupby('Source Id')['Uid'].nunique()
costo_adquisicion_cliente = gastos_por_fuente / clientes_por_fuente

costo_adquisicion_cliente
```

Se calculó el costo de adquirir un nuevo cliente por fuente. La fuente más cara tiene un CAC  de 10.20 y la más barata 1.97

<div class="alert alert-block alert-success"> <b>Comentario del revisor</b> <a class="tocSkip"></a><br>

<b>Éxito</b> - ¡Excelente análisis del costo de adquisición de clientes (CAC) por fuente! Identificar las diferencias entre las fuentes más caras y más económicas proporciona información clave para optimizar las estrategias de marketing. Este tipo de métricas es fundamental para la toma de decisiones basada en datos. ¡Bien hecho! 🎯👏
</div>

### ¿Cuán rentables eran las inversiones? (ROMI)

```python
# Calcular ingresos por fuente
ingresos_por_fuente = pedidos_por_fuente.groupby('Source Id')['Revenue'].sum()

# Calcular ROMI
romi = (ingresos_por_fuente / gastos_por_fuente) - 1
romi
```

Todas las fuentes analizadas son rentables, excepto una. Con ROMIs que van desde 1.09% hasta 109.31%. Lo que sigifica que hay fuentes significativamente más rentables que otras.

<div class="alert alert-block alert-success"> <b>Comentario del revisor</b> <a class="tocSkip"></a><br>

<b>Éxito</b> - ¡Gran trabajo calculando el Retorno de la Inversión en Marketing (ROMI)! Identificar fuentes rentables y destacar aquellas que no lo son permite priorizar esfuerzos y recursos. Además, reconocer la variabilidad en los ROMIs es crucial para ajustar estrategias y maximizar el retorno. ¡Sigue así, este análisis agrega mucho valor! 🚀📊
